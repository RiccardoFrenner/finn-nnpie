```python
import numpy as np
import matplotlib.pyplot as plt
import keras
from keras import Sequential, layers, optimizers

print(keras.version())
```

```python


# Get data
np.random.seed(0)
x_data = np.load("x_res_-1.npy")
x = np.linspace(x_data.min(), x_data.max(), 2000)
y_noisy = np.load("y_res_-1.npy")

# Define the model
model = Sequential(
    [
        layers.Dense(64, input_dim=1, activation="relu"),
        layers.Dense(64, activation="relu"),
        layers.Dense(1),
    ]
)

model.summary()
# Compile the model
model.compile(optimizer=optimizers.Adam(learning_rate=1e-2), loss="mse")

# Train the model
model.fit(
    x_data,
    y_noisy,
    epochs=100,
    batch_size=16,
    validation_split=0.2,
    callbacks=[
        keras.callbacks.EarlyStopping(
            monitor="val_loss",
            patience=50,
            restore_best_weights=True,
            min_delta=1e-3,
        )
    ],
    verbose=1,
)

# Predict using the trained model
y_pred = model.predict(x)

# Plot the data and the learned function
plt.figure(figsize=(10, 6))
plt.scatter(x_data, y_noisy, label="Noisy Data", color="blue", alpha=0.1)
plt.plot(x, y_pred, label="Learned Function", color="red", lw=3)
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

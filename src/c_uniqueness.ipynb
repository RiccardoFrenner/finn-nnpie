# Compute derivatives to compute R(c) from equation
$$
\frac{\partial c}{\partial t} = \frac{D}{R(c)} \frac{\partial^2 c}{\partial x^2}
$$

```python
c_full.shape
```

```python
dcdt = (c_full[2:] - c_full[:-2]) / (2*Nt)
d2cdx2 = (c_full[..., 2:] - 2 * c_full[..., 1:-1] + c_full[..., :-2]) / Nt**2

# dcdt = np.concatenate([np.zeros((1, Nx)), dcdt, np.zeros((1, Nx))], axis=0)
# d2cdx2 = np.concatenate([d2cdx2[:, 0:1], d2cdx2, d2cdx2[:, -1:]], axis=1)
dcdt = dcdt[:, 1:-1]
d2cdx2 = d2cdx2[1:-1, :]
print(dcdt.shape, d2cdx2.shape)
```

```python
fig, axs = plt.subplots(nrows=3, figsize=(12, 12))
axs[0].pcolormesh(t[1:-1], x[1:-1], d2cdx2.T)
axs[1].pcolormesh(t[1:-1], x[1:-1], d2cdx2.T, vmin=d2cdx2.min(), vmax=1e-7)
axs[2].pcolormesh(t[1:-1], x[1:-1], np.log(d2cdx2.T))
plt.show()
```

```python
fig, axs = plt.subplots(nrows=3, figsize=(12, 12))
axs[0].pcolormesh(t[1:-1], x[1:-1], dcdt.T)
axs[1].pcolormesh(t[1:-1], x[1:-1], dcdt.T, vmin=dcdt.min(), vmax=1e-5)
axs[2].pcolormesh(t[1:-1], x[1:-1], np.log(dcdt.T))
plt.show()
```
